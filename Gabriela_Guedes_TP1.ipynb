* **Descrição:** Sistema para adicionar, listar, concluir e remover tarefas.



#    Configuração Inicial



```python
from datetime import datetime

tarefas=[]
contador_id = 1
```

#    Função adicionar_tarefa

```python
"""def===Definir uma função, sempre que começar uma função usar!!!"""
def adicionar_tarefa(descricao, prazo=None, urgencia="Normal"):
  global contador_id

  nova_tarefa = {
      'id':  contador_id,
      'descricao': descricao,
      'data_criacao':  datetime.now().strftime("%d/%m/%Y %H:%M"),
      'prazo': prazo if prazo else "Sem Prazo",
      'urgencia': urgencia,
      'status':"Pendente"
  }

  tarefas.append(nova_tarefa)
  contador_id += 1

  print(f"Tarefa #{nova_tarefa['id']} adicionada!")
  return nova_tarefa
```

#    Função listar_tarefa

```python
def listar_tarefas(filtro_status=None):
  if not tarefas:
    print("Nenhuma tarefa cadastrada")
    return 0

    """list comprehension para compactar a lista de filtragem"""
  if filtro_status:
    lista_exibir = [t for t in tarefas if t["status"] == filtro_status]
  else:
    lista_exibir = tarefas

  if not lista_exibir:
    print(f"Nenhuma tarefa '{filtro_status}' encontrada")
    return 0

  print("\n" + "=" * 50)
  print("LISTA DE TAREFAS")
  print("=" * 50)

  for tarefa in lista_exibir:
    print(f"\n ID #{tarefa['id']} | {tarefa['status'].upper()}| Urgência: {tarefa['urgencia']}")
    print(f"{tarefa['descricao']}")
    print(f"Criada: {tarefa['data_criacao']} | Prazo: {tarefa['prazo']}")
    print(" " + "-" * 46)

  print(f"\n Total: {len(lista_exibir)} tarefa(s) \n")
  return len(lista_exibir)
```

#    Função concluida_tarefa

```python
def concluida_tarefa(id_tarefa):
  for tarefa in tarefas:
    if tarefa["id"] == id_tarefa:
      if tarefa["status"] == "concluída":
        print(f"Tarefa #{id_tarefa} já estava concluída.")
        return False
      tarefa["status"] = "concluída"
      print(f"Tarefa #{id_tarefa} concluída: '{tarefa['descricao']}'")
      return True

  print(f"Tarefa #{id_tarefa} não encontrada")
  return False
```

#    Função remover_tarefa

```python
def remover_tarefa(id_tarefa, confirmar=True):
  for i, tarefa in enumerate(tarefas):
    if tarefa["id"] == id_tarefa:
      if confirmar:
        resposta = input(f"Remover '{tarefa['descricao']}'? (s/n): ").strip().lower()
        if resposta != "s":
          print("Remoção cancelada")
          return False
      tarefas.pop(i)
      print(f"Tarefas #{id_tarefa} removida!")
      return True
  print(f"Tarefa #{id_tarefa} não encontrada.")
  return False
```

#    Menu exibir_tarefa


```python
def exibir_menu():
    print("\n" + "=" * 40)
    print("--GERENCIADOR DE TAREFAS--")
    print("=" * 40)
    print("[1] Adicionar tarefa")
    print("[2] Listar todas")
    print("[3] Listar pendentes")
    print("[4] Listar concluídas")
    print("[5] Marcar como concluída")
    print("[6] Remover tarefa")
    print("[0] Sair")
    print("=" * 40)
    return input("Sua escolha: ").strip()


def main():

    print("Bem-vindo(a)!")

    while True:
        opcao = exibir_menu()

        if opcao == "1":
            descricao = input("Descrição: ").strip()
            if not descricao:
                print("Descrição não pode ser vazia!")
                continue
            prazo = input("Prazo (DD/MM/AAAA) ou Enter para pular: ").strip() or None
            print("Urgência: [1] Baixa  [2] Normal  [3] Alta")
            op = input("Escolha (Enter = Normal): ").strip()
            urgencia = {"1": "baixa", "2": "normal", "3": "alta"}.get(op, "normal")
            adicionar_tarefa(descricao, prazo=prazo, urgencia=urgencia)

        elif opcao == "2":
            listar_tarefas()

        elif opcao == "3":
            listar_tarefas(filtro_status="pendente")

        elif opcao == "4":
            listar_tarefas(filtro_status="concluída")

        elif opcao == "5":
            listar_tarefas(filtro_status="pendente")
            try:
                id_t = int(input("ID da tarefa: "))
                concluida_tarefa(id_t)
            except ValueError:
                print("Digite apenas números")

        elif opcao == "6":
            listar_tarefas()
            try:
                id_t = int(input("ID da tarefa: "))
                remover_tarefa(id_t)
            except ValueError:
                print("Digite apenas números")

        elif opcao == "0":
            print("Fim do uso")
            break

        else:
            print("Opção inválida!")


main()
```
